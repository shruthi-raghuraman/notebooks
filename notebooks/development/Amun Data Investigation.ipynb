```python
def doc_structure(x, lst, lst2, lst3, lst4):
    for i in x.keys():
        if type(x[i]) is not dict:
            print(i, "is not a dictionary")
            lst.append(i)
        else:
            print(i, x[i].keys())
            for j in x[i].keys():
                if type(x[i][j]) is dict:
                    print(j)
                    lst2.append(j)
                    doc_structure(x[i][j], lst, lst2, lst3, lst4)
                else:
                    lst3.append(j)
                    lst4.append(x[i][j])
```
